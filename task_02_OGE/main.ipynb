### 6. Проверка гипотезы: зависимость результатов от удалённости от областного центра.

```python
# склейка по коду Района + Расстояние до Саратова (без Неизвестных районов)
merged = new_oge.merge(
    geo[["Код района", "Расстояние до Саратова"]], 
    on="Код района", 
    how="left"
)
```

```python
# группировка + новые значения по среднему и расстоянию
rayons_hypo = merged.groupby("Район").agg(
    Средний_балл=("% Оценка", "mean"),
    Расстояние_км=("Расстояние до Саратова", "first")
).reset_index()
```

```python
plt.figure(figsize=(12, 6))

sns.regplot(
    data=rayons_hypo,
    x="Расстояние_км",
    y="Средний_балл"
)

plt.title("Зависимость среднего балла от удалённости от Саратова")
plt.xlabel("Расстояние")
plt.ylabel("Средний балл")
plt.grid(True)
plt.show()
```

### Выводы
 - В целом средняя оценка по всем расстояниям до Саратова 60, но чем дальше от Саратова тем ниже линия регрессии
 - Также можно заметить что больше точек находятся выше 70 баллов только в 0 расстоянии
 - Есть выброс в районе 200-250 километров (списали)
